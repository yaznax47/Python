```python
from bs4 import BeautifulSoup
url = BeautifulSoup('https://www.fifa.com/fifa-world-ranking/ranking-table/men/#UEFA', 'html.parser')
url
```

```python
import requests
response = requests.get(url)
response
```

```python
obj=response.text
obj = BeautifulSoup(obj,"lxml")
obj
```

```python
table = obj.find('table', attrs={'id':'rank-table'})
```

```python
table
```

```python
rows = table.find_all("tr")
```

```python
rows
```

```python
tdata = table.find_all("td")
```

```python
data=[]
for i in rows:
    data.append(i.text)
```

```python
data
```

```python
output=[]
for i in range(len(data)):
    x=data[i].split('\n')[1:-1]
    output.append(x)  
```

```python
output
```

```python
finals=[]
ls=[0,7,13,14,15]
for i in output:
    jj=[]
    for j in ls:
        jj.append(i[j])
    finals.append(jj)
finals[0]=['RANK','TEAM','TOTAL POINTS','PREVIOUS POINTS','+/-']    
finals
```

```python
import csv
with open('UEFA.csv','w') as f:
    for i in finals:
        x=csv.writer(f)
        x.writerow(i)
import pandas as pd
df=pd.read_csv('UEFA.csv',encoding='latin',index_col='RANK',nrows=50)
df
```
