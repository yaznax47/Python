```python
from bs4 import BeautifulSoup
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
```

```python
import requests
response = requests.get(url)
response
```

```python
obj=response.text
obj = BeautifulSoup(obj,"lxml")
obj
```

```python
table = obj.table
table
```

```python
tv = table.find_all('tr')
tv
```

```python
row=[]
for i in tv:
    row.append(i.text)
row
```

```python
final_row=[]
for i in row:    
    row1 = i.split('\n')
    final_row.append(row1)
final_row
```

```python
import csv 
with open('corona1.csv','a') as f:
    x = csv.writer(f)
    for i in final_row:
        x.writerow(i)
        print(i)
```
