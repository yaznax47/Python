```python
# function with no argument and return type

def area():
    l = 100
    b = 20
    a = l*b
    return a

ar = area()
print("The area is = ",ar)
h = 10
v = ar*h
print(v)
```

```python
# function with argument and return type

def area(x,y):
    a = x*y
    return a

l = 100
b = 20

ar = area(l,b)
print("The area is = ",ar)
h = 10
v = ar*h
print(v)
```

```python
def area(l,b):
    a = l*b
    return a

def volume(l,b,h):
    v = area(l,b)*h
    return v
    
l = 10
b = 5
h =2
print(area(l,b))
print(volume(l,b,h))
```

```python
# Recursive function
def hello():
    print("Hello World")
    hello()
    
hello()
```

```python
# Recursive function
def hello():
    print("Hello World")
    x = input("Enter y for more print = ")
    if x == 'y':
        hello()
    
hello()
```

```python
# Recursive function
def hello():
    p = float(input("Enter p = "))
    t = float(input("Enter t = "))
    r = float(input("Enter r = "))
    i = p*t*r/100
    print("The value of i = ",i)
    x = input("Enter y for more print = ")
    if x == 'y':
        hello()
    
hello()
```
